## Task 1

```python
menu = {
    "Paneer Tikka":   {"category": "Starters",  "price": 180.0, "available": True},
    "Chicken Wings":  {"category": "Starters",  "price": 220.0, "available": False},
    "Veg Soup":        {"category": "Starters",  "price": 120.0, "available": True},
    "Butter Chicken": {"category": "Mains",     "price": 320.0, "available": True},
    "Dal Tadka":      {"category": "Mains",     "price": 180.0, "available": True},
    "Veg Biryani":    {"category": "Mains",     "price": 250.0, "available": True},
    "Garlic Naan":    {"category": "Mains",     "price":  40.0, "available": True},
    "Gulab Jamun":    {"category": "Desserts",  "price":  90.0, "available": True},
    "Rasgulla":       {"category": "Desserts",  "price":  80.0, "available": True},
    "Ice Cream":      {"category": "Desserts",  "price": 110.0, "available": False},
}

# --- Task 1: Grouped Menu Print ---
# get unique categories to avoid hardcoding headers
categories = sorted(set(info["category"] for info in menu.values()))

for cat in categories:
    print(f"\n===== {cat} =====")
    for item, info in menu.items():
        if info["category"] == cat:
            status = "Available" if info["available"] else "Unavailable"
            # align text for a clean "restaurant bill" look
            print(f"{item:<15} ₹{info['price']:>6.2f}   [{status}]")

# --- Task 2: Dictionary Analytics ---
print("\n" + "="*30)
print("RESTAURANT ANALYTICS")
print("="*30)

# total count of keys
total_items = len(menu)
# filter by boolean flag
available_items = len([i for i in menu.values() if i["available"]])

# find max value based on the 'price' key inside the nested dict
most_expensive = max(menu.items(), key=lambda x: x[1]["price"])
# filter items based on price threshold
budget_items = [(name, info["price"]) for name, info in menu.items() if info["price"] < 150]

print(f"Total Items: {total_items}")
print(f"Available: {available_items}")
print(f"Most Expensive: {most_expensive[0]} (₹{most_expensive[1]['price']})")

print("\nBudget Options (Under ₹150):")
for name, price in budget_items:
    print(f"- {name}: ₹{price}")
```

## Task 2

```python
menu = {
    "Paneer Tikka":   {"category": "Starters",  "price": 180.0, "available": True},
    "Chicken Wings":  {"category": "Starters",  "price": 220.0, "available": False},
    "Veg Soup":        {"category": "Starters",  "price": 120.0, "available": True},
    "Butter Chicken": {"category": "Mains",     "price": 320.0, "available": True},
    "Dal Tadka":      {"category": "Mains",     "price": 180.0, "available": True},
    "Veg Biryani":    {"category": "Mains",     "price": 250.0, "available": True},
    "Garlic Naan":    {"category": "Mains",     "price":  40.0, "available": True},
    "Gulab Jamun":    {"category": "Desserts",  "price":  90.0, "available": True},
    "Rasgulla":       {"category": "Desserts",  "price":  80.0, "available": True},
    "Ice Cream":      {"category": "Desserts",  "price": 110.0, "available": False},
}

cart = []

def add_to_cart(item_name, qty):
    # check if menu contains the item
    if item_name not in menu:
        print(f"❌ '{item_name}' is not on our menu.")
        return
    
    # check if item is currently in stock
    if not menu[item_name]["available"]:
        print(f"❌ Sorry, '{item_name}' is currently unavailable.")
        return

    # check if already in cart to avoid duplicates
    for entry in cart:
        if entry["item"] == item_name:
            entry["quantity"] += qty
            print(f"Updated {item_name} quantity to {entry['quantity']}")
            return
    
    # add new entry if not found in loop
    cart.append({"item": item_name, "quantity": qty, "price": menu[item_name]["price"]})
    print(f"Added {qty}x {item_name} to cart.")

def remove_from_cart(item_name):
    global cart
    # filter out the item to 'remove' it
    initial_len = len(cart)
    cart = [entry for entry in cart if entry["item"] != item_name]
    
    if len(cart) == initial_len:
        print(f"⚠️ '{item_name}' was not in your cart.")
    else:
        print(f"Removed '{item_name}' from cart.")

# --- Simulation Sequence ---

add_to_cart("Paneer Tikka", 2)
add_to_cart("Gulab Jamun", 1)
add_to_cart("Paneer Tikka", 1) # should increment to 3
add_to_cart("Mystery Burger", 1) # missing item
add_to_cart("Chicken Wings", 1) # unavailable item
remove_from_cart("Gulab Jamun")

# --- Final Order Summary ---

print("\n========== Order Summary ==========")
subtotal = 0
for entry in cart:
    item_total = entry["quantity"] * entry["price"]
    subtotal += item_total
    print(f"{entry['item']:<18} x{entry['quantity']}   ₹{item_total:>7.2f}")

gst = subtotal * 0.05
total_payable = subtotal + gst

print("-" * 36)
print(f"{'Subtotal:':<23} ₹{subtotal:>8.2f}")
print(f"{'GST (5%):':<23} ₹{gst:>8.2f}")
print(f"{'Total Payable:':<23} ₹{total_payable:>8.2f}")
print("=" * 36)
```

## Task 3

```python
import copy

inventory = {
    "Paneer Tikka":   {"stock": 10, "reorder_level": 3},
    "Chicken Wings":  {"stock":  8, "reorder_level": 2},
    "Veg Soup":        {"stock": 15, "reorder_level": 5},
    "Butter Chicken": {"stock": 12, "reorder_level": 4},
    "Dal Tadka":      {"stock": 20, "reorder_level": 5},
    "Veg Biryani":    {"stock":  6, "reorder_level": 3},
    "Garlic Naan":    {"stock": 30, "reorder_level": 10},
    "Gulab Jamun":    {"stock":  5, "reorder_level": 2},
    "Rasgulla":       {"stock":  4, "reorder_level": 3},
    "Ice Cream":      {"stock":  7, "reorder_level": 4},
}

# --- Task 1: Deep Copy & Verification ---
# deepcopy clones nested dicts; a normal copy would only clone the outer dict
inventory_backup = copy.deepcopy(inventory)

# temporary change to prove independence
inventory["Paneer Tikka"]["stock"] = 999
print(f"Current Stock (Modified): {inventory['Paneer Tikka']['stock']}")
print(f"Backup Stock (Original): {inventory_backup['Paneer Tikka']['stock']}")

# restore to original state before processing order
inventory = copy.deepcopy(inventory_backup)

# --- Task 2: Order Fulfilment ---
# cart from previous task: [{"item": "Paneer Tikka", "quantity": 3, "price": 180.0}]
print("\n--- Processing Order ---")
for order in cart:
    item = order["item"]
    requested_qty = order["quantity"]
    
    if item in inventory:
        current_stock = inventory[item]["stock"]
        # take what we can, but don't go negative
        deduct_qty = min(requested_qty, current_stock)
        inventory[item]["stock"] -= deduct_qty
        
        if deduct_qty < requested_qty:
            print(f"CAUTION: Insufficient stock for {item}. Only {deduct_qty} processed.")
        else:
            print(f"CAUTION: Deducted {deduct_qty} {item} from inventory.")

# --- Task 3: Reorder Alerts ---
print("\n--- Reorder Alerts ---")
for item, details in inventory.items():
    stock = details["stock"]
    level = details["reorder_level"]
    # check if stock hit the threshold
    if stock <= level:
        print(f"⚠ Reorder Alert: {item} — Only {stock} unit(s) left (reorder level: {level})")

# --- Final Verification ---
print("\nFinal Check:")
# Paneer Tikka should be 7 in inventory, but 10 in backup
print(f"Inventory Paneer: {inventory['Paneer Tikka']['stock']}")
print(f"Backup Paneer:    {inventory_backup['Paneer Tikka']['stock']}")
```

## Task 4

```python
sales_log = {
    "2025-01-01": [
        {"order_id": 1,  "items": ["Paneer Tikka", "Garlic Naan"],          "total": 220.0},
        {"order_id": 2,  "items": ["Gulab Jamun", "Veg Soup"],              "total": 210.0},
        {"order_id": 3,  "items": ["Butter Chicken", "Garlic Naan"],        "total": 360.0},
    ],
    "2025-01-02": [
        {"order_id": 4,  "items": ["Dal Tadka", "Garlic Naan"],             "total": 220.0},
        {"order_id": 5,  "items": ["Veg Biryani", "Gulab Jamun"],           "total": 340.0},
    ],
    "2025-01-03": [
        {"order_id": 6,  "items": ["Paneer Tikka", "Rasgulla"],             "total": 260.0},
        {"order_id": 7,  "items": ["Butter Chicken", "Veg Biryani"],        "total": 570.0},
        {"order_id": 8,  "items": ["Garlic Naan", "Gulab Jamun"],           "total": 130.0},
    ],
    "2025-01-04": [
        {"order_id": 9,  "items": ["Dal Tadka", "Garlic Naan", "Rasgulla"], "total": 300.0},
        {"order_id": 10, "items": ["Paneer Tikka", "Gulab Jamun"],          "total": 270.0},
    ],
}

# --- Task 1: Append New Data ---
sales_log["2025-01-05"] = [
    {"order_id": 11, "items": ["Butter Chicken", "Gulab Jamun", "Garlic Naan"], "total": 490.0},
    {"order_id": 12, "items": ["Paneer Tikka", "Rasgulla"],                     "total": 260.0},
]

# --- Task 2: Daily Revenue & Best Day ---
daily_revenue = {}
item_frequency = {}

for date, orders in sales_log.items():
    # sum up totals for each specific date
    day_total = sum(order["total"] for order in orders)
    daily_revenue[date] = day_total
    
    # track item occurrences across all orders
    for order in orders:
        for item in order["items"]:
            item_frequency[item] = item_frequency.get(item, 0) + 1

print("Daily Revenue Report:")
for date, rev in daily_revenue.items():
    print(f"{date}: ₹{rev:.2f}")

# find key with max value in revenue dict
best_day = max(daily_revenue, key=daily_revenue.get)
# find key with max value in frequency dict
most_ordered = max(item_frequency, key=item_frequency.get)

print(f"\nBest-Selling Day: {best_day} (₹{daily_revenue[best_day]:.2f})")
print(f"Most Ordered Item: {most_ordered} ({item_frequency[most_ordered]} times)")

# --- Task 3: Numbered Master Log ---
print("\n--- Master Sales Log ---")
all_orders = []
# flatten the dict into a single list for easy enumeration
for date, orders in sales_log.items():
    for order in orders:
        all_orders.append((date, order))

for i, (date, order) in enumerate(all_orders, 1):
    items_str = ", ".join(order["items"])
    # clean formatting for a professional audit trail
    print(f"{i}. [{date}] Order #{order['order_id']:<2} — ₹{order['total']:>6.2f} — Items: {items_str}")
```
